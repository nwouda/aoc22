```python
input = open('3.input', 'r')
scored = 0
def get_overlap(pocket1, pocket2):
    for letter in pocket1:
        if letter in pocket2:
            return letter
for bag in input.readlines():
    pocket1 = bag[0:int(len(bag)/2)]
    pocket2 = bag[int(len(bag)/2):len(bag)]
    score = ord(get_overlap(pocket1, pocket2))
    if score > 90:
        score -= 96
    else:
        score -= 38
    scored += score
    print(scored)


```

```python
input = open('3.input', 'r')
scored = 0
def get_overlap(bag1, bag2, bag3):
    for letter in bag1:
        if letter in bag2 and letter in bag3:
            return letter
bags = list(input.readlines())
while len(bags)>0:
    bag = [bags.pop(0), bags.pop(0), bags.pop(0)]
    bag1 = bag[0]
    bag2 = bag[1]
    bag3 = bag[2]
    # print(get_overlap(bag1, bag2, bag3))
    score = ord(get_overlap(bag1, bag2, bag3))
    if score > 90:
        score -= 96
    else:
        score -= 38
    scored += score

print(scored)

```
