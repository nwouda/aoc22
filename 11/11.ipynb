```python
input = open('11.input', 'r')
from math import lcm

monkeys = {}

monkeys_raw = []

monkey_config = []

for line in input.readlines():
    line = line.rstrip('\n')
    # print(line)
    if line != '':
        # print('NOT EMPTY')
        monkey_config.append(line.lstrip())
    else:
        # print('EMPTY')
        monkeys_raw.append(monkey_config)
        monkey_config = []

from pprint import pprint

# pprint(monkeys_raw)

class Monkey:
    mod = 0
    def __init__(
        self,
        id: int,
        items: list,
        operation: str,
        test: str,
        if_true: str,
        if_false: str
    ):
        self.id = id
        self.items = items
        self.operation = operation
        self.test = int(test)
        self.if_true = int(if_true)
        self.if_false = int(if_false)
        self.inspections = 0
        
    @property
    def operation(self) -> str:
        return self._operation

    @operation.setter
    def operation(self, operation):
        op = operation.split(' ')
        self._operation = {
            'a': op[2],
            'act': op[3],
            'b': op[4],
        }

    def _value_item(self, item):
        # print(f'  Monkey inspects an item with a worry level of {item}')
        operation = {}
        if self.operation['a'] == 'old':
            operation['a'] = item
        else:
            operation['a'] = self.operation['a']
        if self.operation['b'] == 'old':
            operation['b'] = item
        else:
            operation['b'] = self.operation['b']
        res = eval(f'{operation["a"]} {self.operation["act"]} {operation["b"]}')
        # print(f'    Worry level increases from {item} to {res}')
        # print(f'    Monkey gets bored with item. Worry level is divided by 3 to {int(res/3)}')
        return int(res/3)

    def _value_item_p2(self, item):
        # print(f'  Monkey inspects an item with a worry level of {item}')
        operation = {}
        if self.operation['a'] == 'old':
            operation['a'] = item
        else:
            operation['a'] = self.operation['a']
        if self.operation['b'] == 'old':
            operation['b'] = item
        else:
            operation['b'] = self.operation['b']
        res = eval(f'{operation["a"]} {self.operation["act"]} {operation["b"]}')
        # print(f'    Worry level increases from {item} to {res}')
        # print(f'    Monkey gets bored with item. Worry level is NOT divided by 3 to {int(res/3)}')
        return res

    def inspect_items(self):
        for i in self.items:
            self.inspections += 1
            value = self._value_item(i)
            if value%self.test > 0:
                # print(f'    Current worry level is not divisible by {self.test}')
                # print(f'    Item with worry level {value} is thrown to monkey {self.if_false}')
                monkeys[self.if_false].add_item(value)
            if value%self.test == 0:
                # print(f'    Current worry level divisible by {self.test}')
                # print(f'    Item with worry level {value} is thrown to monkey {self.if_true}')
                monkeys[self.if_true].add_item(value)
        self.items = []

    def inspect_items_p2(self):
        for i in self.items:
            self.inspections += 1
            value = self._value_item_p2(i)
            if value%self.test > 0:
                # print(f'    Current worry level is not divisible by {self.test}')
                # print(f'    Item with worry level {value} is thrown to monkey {self.if_false}')
                monkeys[self.if_false].add_item(value)
            if value%self.test == 0:
                # print(f'    Current worry level {value} divisible by {self.test}')
                # print(f'    Item with worry level {value} is thrown to monkey {self.if_true}')
                monkeys[self.if_true].add_item(value%self.mod)

        self.items = []

    def add_item(self, item):
        self.items.append(item)
```

```python
mod = 1
for monkey in monkeys_raw:
    num = int(monkey[0].split(' ')[1].split(':')[0])
    items = [int(i) for i in monkey[1].split(':')[1].lstrip().split(', ')]
    print(f'ITEMS: {items}')
    op = monkey[2].split(':')[1].lstrip()
    test = monkey[3].split(' ')[-1]
    mod = mod*int(test)
    if_true = monkey[4].split(' ')[-1]
    if_false = monkey[5].split(' ')[-1]
    monkeys[num] = Monkey(
        num,
        items,
        op,
        test,
        if_true,
        if_false
    )
for monkey in monkeys:
    monkeys[monkey].mod = mod

print(len(monkeys))
```

```python
for i in range(0, 10000):
    if i in [1, 20, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 1000]:
        print(f'== After round {i} ==')
        for m in monkeys:
            print(f'Monkey {m} inspected items {monkeys[m].inspections} times.')
    for m in monkeys:
        # if i%100 == 0:
            # print(f'Monkey {m} has {monkeys[m].items} items.')
        # print(f'Monkey {m}')
        monkeys[m].inspect_items_p2()
```
