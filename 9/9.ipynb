```python
def get_distance(a, b):
    c = a-b
    if c < 0:
        c = 0-c
    return c

def move_rope(h, t):
    # print(h, t)
    for i, ta in enumerate(t):
        # print(f'Processing tail piece {i}')
        if i == 0:
            t[0] = get_direction(h, ta)
            # print(f'moving first tailpiece from {ta} to {t[0]}')
        else:
            t[i] = get_direction(t[i-1], ta)
            # print(f'moving {i}nth tailpiece from {ta} to {t[i]}')
    return t

def get_direction(h, t):
    hdist = get_distance(h['x'], t['x'])
    vdist = get_distance(h['y'], t['y'])
    # print(f'hdist = {hdist}, vdist = {vdist}')
    if hdist == 1 and vdist == 1:
        # print('DO NOTHING')
        return t
    elif hdist > 0 and vdist > 0:
        # print('SHOULD MOVE DIAGONALLY')
        # print(f'HORIZONTAL {h["x"]-t["x"]}')
        dt = t
        if h["x"] > t["x"]:
            # print('Moving tail right')
            dt = { 'x': dt['x']+1, 'y': dt['y'] }
        else:
            # print('Moving tail left')
            dt = { 'x': dt['x']-1, 'y': dt['y'] }
        # print(f'VERTICAL {h["y"]-t["y"]}')
        if h["y"] > t["y"]:
            # print('Moving tail up')
            dt = { 'x': dt['x'], 'y': dt['y']+1 }
        else:
            # print('Moving tail down')
            dt = { 'x': dt['x'], 'y': dt['y']-1 }
        return dt
    elif hdist > 1 or vdist > 1:
        if hdist > vdist:
            # print(f'HORIZONTAL {h["x"]-t["x"]}')
            if h["x"] > t["x"]:
                # print('Moving tail right')
                return { 'x': t['x']+1, 'y': t['y'] }
            else:
                # print('Moving tail left')
                return { 'x': t['x']-1, 'y': t['y'] }
        else:
            # print(f'VERTICAL {h["y"]-t["y"]}')
            if h["y"] > t["y"]:
                # print('Moving tail up')
                return { 'x': t['x'], 'y': t['y']+1 }
            else:
                # print('Moving tail down')
                return { 'x': t['x'], 'y': t['y']-1 }
    else:
        # print('Not enough distance to move')
        return t
```
