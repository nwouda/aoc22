```python
input = open('7.input', 'r')

cwd = []
folders = {}

for line in input.readlines():
    line = line.rstrip('\n')
    command = line.split(' ')
    if command[0] != '$':
        path = str.join('/', cwd)
        if not folders.get(path):
            folders[path] = {}
            folders[path]['size'] = 0
            folders[path]['subdirs'] = []
            folders[path]['path'] = path
        if command[0] == 'dir':
            print(f'found dir {command[1]} at {str.join("/", cwd)}')
            folders[path]['subdirs'].append(command[1])
        else:
            size = command[0]
            folders[path]['size'] += int(size)
            print(f'Folder {path} size: {folders[path]["size"]}')
    elif command[0] == '$':
        if command[1] == 'cd':
            if command[2] == '..':
                print(f'Moving up from {cwd}')
                cwd.pop()
                print(f'Moved to {cwd}')
            else:
                print(f'Entering directoring {command[2]}')
                cwd.append(command[2])
                print(f'Currently in {cwd}')
        elif command[1] == 'ls':
            print(f'Reading dir {cwd}')
    print("EOL\n")

```

```python
final_size = {}
from pprint import pprint
tempsize=[]
def get_subdirs(dir):
    print(f'Starting subdir process for {dir}: {folders[dir]["size"]}')
    if len(folders[dir]['subdirs']) == 0:
        print(f'No subfolders in {dir}, setting size to {folders[dir]["size"]}')
        folders[dir]['final_size'] = folders[dir]['size']
    else:
        tempsize = []
        for d in folders[dir]['subdirs']:
            cw = folders[dir]['path']+'/'+d
            print(cw)
            get_subdirs(cw)
            tempsize.append(int(folders[cw]['final_size']))
        folders[dir]['final_size'] = sum(tempsize, int(folders[dir]['size']))
        print(dir)
        print(folders[dir]['size'])
        print(f'{tempsize} + {folders[dir]["size"]}')
        print(f'final size for {dir}: {folders[dir]["final_size"]}')
        

```
